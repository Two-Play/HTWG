### Custom Verteilung

Verteilung, die durch eine Tabelle dargestellt wird.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def value_rounded_str(value):
    val = str(np.round(value,4))
    return val
```

```python
table = [(2, 0.1), (3, 0.2), (4, 0.4), (5, 0.2), (6, 0.1)]
table = [(-3, 3/12), (-2, 2/12), (-1, 1/12), (0, 0/12), (1, 1/12), (2, 2/12), (3, 3/12)]
```

```python
values, relative_frequencies = zip(*table)
cumulative_probabilities = [sum(relative_frequencies[:i+1]) for i in range(len(relative_frequencies))]

plt.bar(values, relative_frequencies, color='blue', alpha=0.7)
plt.title('Absolute Häufigkeit - Verteilung')
plt.xlabel('Werte')
plt.ylabel('Häufigkeit')
plt.show()

plt.step(values, cumulative_probabilities, where='post', color='purple', alpha=0.7)
plt.title('Relative Häufigkeit - Verteilungsfunktion (CDF)')
plt.xlabel('Werte')
plt.ylabel('Relative Häufigkeit')
plt.show()
```

```python
# Erwartungswert

expect = 0
for e in table:
    expect += e[0] * e[1]
print(f"Erwartungswert: {value_rounded_str(expect)}")
```

```python
# Varianz

var = 0
for e in table:
    var += (e[0] - expect) ** 2 * e[1]
print(f"Varianz: {value_rounded_str(var)}")
print(f"Standardabweichung: {value_rounded_str(np.sqrt(var))}")
```

```python
random_var = []
for e in table:
    random_var.append(e[0])

np.mean(random_var)
print(f"Durchschnitt: {value_rounded_str(np.mean(random_var))}")
```
