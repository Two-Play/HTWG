```python
import statistics
import numpy as np
import matplotlib.pyplot as plt
```

```python
def value_rounded_str(value):
    val = str(np.round(value,4))
    return val
```

```python
# Data
data1 = np.array([
    [151, 100, 135, 107, 156, 72, 88, 96, 120]
])
```

```python
flat_data = data1.flatten()
sorted_data = np.sort(flat_data)
sorted_data = sorted_data.reshape(data1.shape)

unique_values, counts = np.unique(sorted_data, return_counts=True)
absolute_frequency = dict(zip(unique_values, counts))

total_elements = sorted_data.size
relative_frequency = {value.item(): np.round(count / total_elements,4) for value, count in absolute_frequency.items()}
cdf_values = np.cumsum(counts) / sorted_data.size

print("Absolute Häufigkeit:", absolute_frequency)
print("Relative Häufigkeit:", relative_frequency)
```

```python
plt.bar(absolute_frequency.keys(), absolute_frequency.values(), color='blue', alpha=0.7)
plt.title('Absolute Häufigkeit - Verteilung')
plt.xlabel('Werte')
plt.ylabel('Häufigkeit')
plt.show()

plt.step(unique_values, cdf_values, where='post', color='purple', alpha=0.7)
plt.title('Relative Häufigkeit - Verteilungsfunktion (CDF)')
plt.xlabel('Werte')
plt.ylabel('Relative Häufigkeit')
plt.show()
```

```python
# Modalwert
modal = statistics.multimode(data1[0,:])
print("Modalwert: " + str(modal))
```

```python
# Arithmetisches Mittel
mean = np.mean(data1[0,:])
print("Arithmetisches Mittel: " + value_rounded_str(mean))
```

```python
# Median
mean = np.median(data1[0,:])
print("Median: " + value_rounded_str(mean))
```

```python
# Quartile
quantile_025 = np.quantile(data1[0,:], 0.25)
quantile_050 = np.quantile(data1[0,:], 0.50)
quantile_075 = np.quantile(data1[0,:], 0.75)
print("25%-Quantil: " + value_rounded_str(quantile_025))
print("50%-Quantil: " + value_rounded_str(quantile_050))
print("75%-Quantil: " + value_rounded_str(quantile_075))
```

```python
# Interquartilsabstand
iqr = quantile_075 - quantile_025
print("Interquartilsabstand: " + value_rounded_str(iqr))
```

```python
# Spez. Quantil
p_quantile = 0.90
quantile_spec = np.quantile(data1[0,:], p_quantile)
print(str(p_quantile * 100)+ "%-Quantil: " + value_rounded_str(quantile_spec))
```

```python
# Spannweite
max_value = np.max(data1[0,:])
min_value = np.min(data1[0,:])
spannweite = max_value - min_value
print("Min: " + value_rounded_str(min_value))
print("Max: " + value_rounded_str(max_value))
print("Spannweite: " + value_rounded_str(spannweite))
```

```python
# Varianz
var = np.var(data1[0,:],ddof=1)
print("Varianz: " + value_rounded_str(var))
```

```python
# Empirische Standardabweichung
std = np.std(data1[0,:],ddof=1)
print("Empirische Standardabweichung: " + value_rounded_str(std))
```
