```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
# Define the parameters for W and E
mu_plot1 = 20  # mean of plot1
sigma_plot1 = 10  # standard deviation of plot1
mu_plot2 = 500  # mean of plot2
sigma_plot2 = 100  # standard deviation of plot2
```

```python
# Define the range of x for the normal distributions
x_w = np.linspace(mu_plot1 - 4 * sigma_plot1, mu_plot1 + 4 * sigma_plot1, 1000)
x_e = np.linspace(mu_plot2 - 4 * sigma_plot2, mu_plot2 + 4 * sigma_plot2, 1000)

# Calculate the normal distribution for W and E
pdf_w = norm.pdf(x_w, mu_plot1, sigma_plot1)
pdf_e = norm.pdf(x_e, mu_plot2, sigma_plot2)

# Plotting the probability density functions for W and E
plt.figure(figsize=(10, 6))
plt.plot(x_w, pdf_w, label='plot1', color='magenta')
plt.plot(x_e, pdf_e, label='plot2', color='yellow')

# Labeling the axes and the plot
plt.title('Wahrscheinlichkeitsdichten von plot1 und plot2')
plt.xlabel('X')
plt.ylabel('Wahrscheinlichkeitsdichte')
plt.legend()
plt.grid(True)
plt.show()
```
