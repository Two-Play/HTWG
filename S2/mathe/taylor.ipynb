```python
from sympy import *
import sympy as sp
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import display, Latex


init_printing(use_latex=True)
```

```python
def calculate_taylor_polynomial(function, degree, expansion_point):
    taylor_polynomial = 0
    x = symbols('x')

    for n in range(degree + 1):
        derivative = function.diff(x, n)
        derivative_value = derivative.subs(x, expansion_point)
        term = (derivative_value / factorial(n)) * (x - expansion_point)**n
        taylor_polynomial += term

    return taylor_polynomial
```

```python
def plot_taylor(label, x_values, y_values):
    plt.figure(figsize=(10, 6))
    plt.plot(x_values, y_values, label=label)
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.title('Plot of the function f(x) and its Taylor polynomial')
    plt.xlim(-10, 10)
    plt.ylim(-10, 10)
    plt.xticks(np.arange(-10, 2, 1))
    plt.yticks(np.arange(-10, 6, 1))

    plt.axhline(0, color='black', lw=1)
    plt.axvline(0, color='black', lw=1)
    plt.legend(loc='best')
    plt.grid(True)
    plt.show()

```
